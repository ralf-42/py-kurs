# **2 <font color='orange'>|</font> Überblick Visuals**
---

Plotly Express (px) ist ein integrierter Teil der Plotlylibrary und der empfohlene Ausgangspunkt für die Erstellung der gängigsten Figuren. Jede Funktion von Plotly Express verwendet intern Diagrammobjekte und gibt eine Abbildungsinstanz zurück. Plotly Express bietet mehr als 30 Funktionen zum Erstellen verschiedener Arten von Zahlen, von einem Streudiagramm über ein Balkendiagramm bis hin zu einem Histogramm und einem Sunburst-Diagramm während einer Datenexplorationssitzung.

| Type             |  Visual                                                         |
|------------------|-----------------------------------------------------------------|
| Basics           |  scatter, line, area, bar, funnel, timeline                     |
| Part-of-Whole    |  pie, sunburst, treemap, funnel_area                            |
| 1D Distributions |  histogram, box, violin, strip                                  |
| 2D Distributions |  density_heatmap, density_contour                               |
| Matrix Input     |  imshow                                                         |
| 3-Dimensional    |  scatter_3d, line_3d                                            |
| Multidimensional |  scatter_matrix, parallel_coordinates, parallel_categories      |
| Tile Maps        |  scatter_mapbox, line_mapbox, choropleth_mapbox, density_mapbox |
| Outline Maps     |  scatter_geo, line_geo, choropleth                              |
| Polar Charts     |  scatter_polar, line_polar, bar_polar                           |
| Ternary Charts   |  scatter_ternary, line_ternary                                  |


# **3 <font color='orange'>|</font> Import und Basisdaten**
---


```
import plotly.express as px

tips = px.data.tips()
gapminder = px.data.gapminder()
oceania = px.data.gapminder().query("continent=='Oceania'")
iris = px.data.iris()
```

# **4 <font color='orange'>|</font> Scatter**
---

[Beispiele](https://plotly.com/python/line-and-scatter/)

<p><font color='black' size="5">
Simple Scatter Chart
</font></p>

```
fig = px.scatter(tips,
                 x="total_bill",
                 y="tip",
                 title = "Analyse Trinkgeld",
                 labels={"total_bill": "Rechnung","tip": "Trinkgeld"},
                 width=800, height=400)
fig.show()
```

<p><font color='black' size="5">
Scatter and Trendline
</font></p>


[Beispiele](https://plotly.com/python/linear-fits/)

```
fig = px.scatter(tips,
                 x="total_bill",
                 y="tip",
                 trendline="ols",
                 trendline_color_override="red",
                 width=800, height=400)
fig.show()
```


<p><font color='black' size="5">
Scatter, Line and Marginals
</font></p>

```
fig = px.scatter(tips,
                 x="total_bill",
                 y="tip",
                 color="time",
                 marginal_y="box",
                 marginal_x="box",
                 trendline="ols",
                 width=1000, height=500)
fig.show()
```

<p><font color='black' size="5">
Scatter Matrix
</font></p>

```
fig = px.scatter_matrix(tips,
                        color="time",
                        width=1000,
                        height=1000)
fig.show()
```

<p><font color='black' size="5">
Scatter Animated
</font></p>


```
px.scatter(gapminder[gapminder['continent'] == 'Europe'],
           x='lifeExp',
           range_x=[40,90],
           y="gdpPercap",
           range_y=[800,42000],
           color='country',
           size ='pop',
           title = "Lebenserwartung im Zeitablauf",
           labels={"year": "Jahr","lifeExp": "Lebenserwartung", "gdpPercap": "Bruttoinlandsprodukt (BIP) pro Kopf "},
           animation_frame="year",
           width=1200,
           height=600)
```

# **5 <font color='orange'>|</font> Line**
---

[Beispiele](https://plotly.com/python/line-charts/)

```
fig = px.line(oceania,
              x="year",
              y="lifeExp",
              color='country',
              width=1200,
              height=600)
fig.show()
```

# **9 <font color='orange'>|</font> Funnel**
---




[Beispiele](https://plotly.com/python/funnel-charts/)

```
funnel = dict(
    Stage=["Visit website", "View Products", "Add to basket", "Checkout"],
    Values=[100, 43, 25, 2])

fig = px.funnel(funnel,
                x='Values',
                y='Stage',
                title='E-Commerce Funnel',
                width=800,
                height=500)
fig.show()
```

# **10 <font color='orange'>|</font> Waterfall**
---




[Beispiele](https://plotly.com/python/waterfall-charts/)

```
import plotly.graph_objects as go

steps = ["2021", "Sales", "Consulting", "Net revenue", "Purchases", "Other expenses", "2022"]
values = [100, 60, 80, 0, -40, -60, 0]
position = ["absolute", "relative","relative", "total", "relative", "relative", "total"]
text_ = ["Total", "+60", "+80", "", "-40", "-60", "Total"]

fig = go.Figure(go.Waterfall(
    name = "Company A",
    orientation = "v",
    measure = position,
    x = steps,
    y = values,
    base = 0,
    textposition = "outside",
    text = text_,
    connector = {"line":{"color":"black"}},
))

fig.update_layout(
        title = "P&L Reconciliation",
        width=1000,
        height=500,
        showlegend = True)

fig.show()
```
