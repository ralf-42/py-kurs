Pythons Syntax wurde ab Version 3.10 um die ``match...case``-Struktur erweitert. Das Structural-Pattern-Matching erm√∂glicht es, eine Variable mit einem Satz m√∂glicher Werte abzugleichen. match...case fungiert in seiner einfachsten Form quasi als Ersatz f√ºr eine if-else-Struktur.

Grunds√§tzlicher Aufbau:

````
match variable:
    case 1:
        Anweisung 1
        Anweisung 2
    case 2:
        Anweisung 3
        Anweisung 4  
    case _:                 # kein match
        Anweisung 5  
        Anweisung 6

````





`````python
# Beispiel: einfaches Matching
#
cmd = input("Befehl eingeben: ")
match cmd:
    case "quit":
        print(cmd)
    case "load":
        print(cmd)
    case "save":
        print( cmd)
    case _:
        print (f"Befehl '{cmd}' nicht verstanden")
`````

Bei einer Schleife/Iteration wird ein Code-Block mehrfach durchgef√ºhrt. Die Anzahl der Wiederholungen kann auf zwei Arten festgelegt werden:

## 7.1 <font color='red'>|</font> Ereignisgesteuerte Schleifen
---



Bei ereignisgesteurten Schleifen/Iterationen wird ein Code-Block solange wiederholt bis eine Bedingung erf√ºllt ist. Ist die Bedingung von Anfang an nicht erf√ºllt, dann wird dieser Code-Block nicht ausgef√ºhrt/√ºbersprungen.

Grunds√§tzlicher Aufbau:

```
        while Bedingung:
            Anweisung 1
            Anweisung 2
        else:
            Anweisung 3
            Anweisung 4
        Anweisung x
        Anweisung y
```


`````python
# Beispiel: Gib i aus, solange i kleiner als 6 ist:
#
i = 1
while i < 6:
  print(i)
  i += 1
else:
  print("Bedingung nicht mehr erf√ºllt, i =", i)

print("Ende")
`````


4! = 1 * 2 * 3 * 4

`````python
class Auto:
  def __init__(self, hersteller, modell, leistung): # Methode zur Erstellung des Blaupause / Bauplan der Klasse
    self.hersteller = hersteller
    self.modell = modell
    self.leistung = leistung
    self.energie = 100
    self.x_pos = 0
    self.y_pos = 0

  def __str__(self):
    return f"{self.hersteller}, {self.modell}, {self.leistung}"

class Antrieb:
  def __init__(self, kraftstoff, antriebsart): # Methode zur Erstellung des Blaupause / Bauplan der Klasse
    self.kraftstoff = kraftstoff
    self.antriebsart = antriebsart

  def __str__(self):
    return f"{self.kraftstoff} {self.antriebsart}"

class Kompakt(Auto, Antrieb):
  def __init__(self, hersteller, modell, leistung, kraftstoff, antriebsart):
    Auto.__init__(self, hersteller, modell, leistung)
    Antrieb.__init__(self, kraftstoff, antriebsart)

  def __str__(self):
    return f"{self.hersteller}, {self.modell}, {self.leistung}, {self.kraftstoff} {self.antriebsart}"
`````

Wird eine Variable mit demselben Namen innerhalb einer Funktion erstellen, ist diese Variable lokal und kann nur innerhalb der Funktion verwendet werden. Bei Aufruf des Code-Blocks vergibt Python einen internen Namen. Wird der Code-Block als eigenst√§ndiges Programm aufgerufen, ist dieser Name ``__main__``. Ein Thread ist ein sequentieller Abarbeitungslauf innerhalb eines Prozesses und teilt sich mit den anderen vorhandenen Threads (Multithreading) des zugeh√∂rigen Prozesses eine Reihe von Betriebsmitteln. Threads innerhalb des gleichen Prozesses k√∂nnen verschiedenen Prozessoren zugeordnet sein.


`````python
# Beispiel: Threading
#
import math
import time
from threading import Thread

def tu_was(i, j):
    time.sleep(1)
    _ = math.sqrt(i**j)
    return ((i+1)*(j+10))

def funktion_1():
    i = 0
    for j in range(10):
        liste.append(tu_was(i, j))

def funktion_2():
    i = 1
    for j in range(10):
        liste.append(tu_was(i, j))


liste = list()

thread1 = Thread(target=funktion_1)
thread2 = Thread(target=funktion_2)

thread1.start()
thread2.start()

thread1.join()  # warte bie alle threads abgeschlossen sind
thread2.join()  # warte bie alle threads abgeschlossen sind

print(liste)
`````

## 17.3 <font color='red'>|</font> Multiprocessing

Multiprocessing ist die Verwendung von zwei oder mehr zentralen Verarbeitungseinheiten (CPUs) innerhalb eines einzelnen Computersystems . Der Begriff bezieht sich auch auf die F√§higkeit eines Systems, mehr als einen Prozessor zu unterst√ºtzen oder Aufgaben zwischen ihnen zuzuweisen.


`````python
# Beispiel: Mehrere Prozesses laufen parallel - Variation Anzahl der Prozesse √ºber n_jobs
#
from joblib import Parallel, delayed
import math
import time

def tu_was(i, j):
    time.sleep(1)
    _ = math.sqrt(i**j)
    return ((i+1)*(j+10))

liste = list()

liste = Parallel(n_jobs=4)(delayed(tu_was)(i, j) for i in range(2) for j in range(10))

print(liste)
`````

`````python
# Beispiel: iterative Funktion
#
def fib_i(n):
    a, b = 0, 1
    for i in range(n):
        a, b = b, a + b
    return a

print(fib_i(30))
`````

`````python
# Beispiel: rekursive Funktion
#
def fib_r(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fib_r(n-1) + fib_r(n-2)

print(fib_r(30))
`````

`````python
# Beispiel: rekursive Funktion - Version 2 mit Memo
#
memo = {0:0, 1:1}
def fib_m(n):
    if not n in memo:
        memo[n] = fib_m(n-1) + fib_m(n-2)
    return memo[n]

print(fib_m(30))
`````

`````python
#
# Datei hochladen
#
from google.colab import files

uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{file_name}" with length {length} bytes'.format(
      file_name=fn, length=len(uploaded[fn])))
file_name = fn
`````

`````python
#
# Version 1 - exklusive doc-strings
#
# reading the file
with open(file_name) as fp:
    contents=fp.readlines()

# initialize two counter to check mismatch between "(" and ")"
open_bracket_counter=0
close_bracket_counter=0

# whenever an element deleted from the list length of the list will be decreased
decreasing_counter=0

for number in range(len(contents)):

    # checking if the line contains "#" or not
    if "#" in contents[number-decreasing_counter]:

        # delete the line if startswith "#"
        if contents[number-decreasing_counter].startswith("#"):
            contents.remove(contents[number-decreasing_counter])
            decreasing_counter+=1

        # delete the character after the "#"
        else:
            newline=""
            for character in contents[number-decreasing_counter]:
                if character=="(":
                    open_bracket_counter+=1
                    newline+=character
                elif character==")":
                    close_bracket_counter+=1
                    newline+=character
                elif character=="#" and open_bracket_counter==close_bracket_counter:
                    break
                else:
                    newline+=character
            contents.remove(contents[number-decreasing_counter])
            contents.insert(number-decreasing_counter,newline)

new_file_name = "new_" + file_name
# writing into a new file
with open(new_file_name,"w") as fp:
    fp.writelines(contents)
`````

`````python
#
# Version 2 - inklusive doc-strings
#
import io, tokenize, re
def remove_comments_and_docstrings(source):
    io_obj = io.StringIO(source)
    out = ""
    prev_toktype = tokenize.INDENT
    last_lineno = -1
    last_col = 0
    for tok in tokenize.generate_tokens(io_obj.readline):
        token_type = tok[0]
        token_string = tok[1]
        start_line, start_col = tok[2]
        end_line, end_col = tok[3]
        ltext = tok[4]
        if start_line > last_lineno:
            last_col = 0
        if start_col > last_col:
            out += (" " * (start_col - last_col))
        if token_type == tokenize.COMMENT:
            pass
        elif token_type == tokenize.STRING:
            if prev_toktype != tokenize.INDENT:
                if prev_toktype != tokenize.NEWLINE:
                    if start_col > 0:
                        out += token_string
        else:
            out += token_string
        prev_toktype = token_type
        last_col = end_col
        last_lineno = end_line
    out = '\n'.join(l for l in out.splitlines() if l.strip())
    return out


# reading the file
with open(file_name) as f_in:
  new_file_name = "new_" + file_name
  with open(new_file_name,"w") as f_out:
    f_out.writelines(remove_comments_and_docstrings(f_in.read()))
`````
