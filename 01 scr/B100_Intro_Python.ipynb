Python akzeptiert auch Funktionsrekursion, was bedeutet, dass eine definierte Funktion sich selbst aufrufen kann.

4! = 1 * 2 * 3 * 4

Quelle: Wikipedia

## <b>10.9 <font color='red'>|</font>  Method Resolution Order
---


In Python k√∂nnen Konflikte durch die Methodenaufl√∂sungsreihenfolge (MRO, Method Resolution Order) gel√∂st.

`````python
class A:
    def __init__(self, f1, f2):
        self.feld1 = f1
        self.feld2 = f2

class B():
    def __init__(self, f1):
        self.feld1 = f1
`````

`````python
# _init__ von Klasse A wird aufgerufen, __init__ von Klasse B entf√§llt
class C(A, B):
    pass

obj = C("A", "a")
obj.feld1, obj.feld2
`````

`````python
# _init__ von Klasse B wird aufgerufen, __init__ von Klasse A entf√§llt
class C(B, A):
    pass

obj = C("B")
obj.feld1
`````

`````python
help(C)
`````

Der Methodenaufl√∂sungsmechanismus (MRO) kann umgangen werden, wenn die Initialisierer beider Superklassen (A und B) explizit aufgerufen haben.

`````python
class A:
    def __init__(self, f1, f2):
        self.feld1 = f1
        self.feld2 = f2
class B():
    def __init__(self, f1):
        self.feld1 = f1
`````

`````python
class C(B, A):
  def __init__(self, f1, f2):
    B.__init__(self, f1)
    A.__init__(self, f1, f2)
`````

`````python
obj = C("A", "B")
`````

`````python
obj.feld1, obj.feld2
`````

`````python
help(C)
`````

## <b>10.10 <font color='red'>|</font>  Programmierparadigmen
---

| Aspekt      | Funktionale Programmierung      | Objektorientierte Programmierung |
|-------------|---------------------------------|----------------------------------|
| **Vorteile**    |                                 |                                  |
| Einfachheit | Oft k√ºrzerer und klarerer Code  | Klare Struktur durch Nutzung von Klassen und Objekten |
| Zustand     | Kein ver√§nderbarer Zustand, was zu weniger Seiteneffekten f√ºhren kann | Verwaltung von Zust√§nden und Datenkapselung ist oft einfacher |
| Wiederverwendbarkeit | H√∂here Modularit√§t und Wiederverwendbarkeit durch reine Funktionen | Wiederverwendbarkeit durch Erbschaft und Polymorphismus |
| Concurrency | Oft einfacher f√ºr parallele Verarbeitung, da keine Zustands√§nderungen | Kann komplizierter sein aufgrund der Notwendigkeit, Zustands√§nderungen zu synchronisieren |
| **Nachteile**   |                                 |                                  |
| Lernkurve   | Kann unintuitiv sein f√ºr Personen, die mit imperativer Programmierung vertraut sind | Kann komplex werden, wenn Klassen und Vererbungshierarchien gro√ü sind |
| Leistung   | Kann in einigen F√§llen weniger effizient sein | Speicheroverhead durch Objektinstanzen |
| Design     | Erfordert oft ein Umdenken bei der Probleml√∂sung | Kann zu √ºberm√§√üiger Komplexit√§t f√ºhren, wenn nicht sorgf√§ltig entworfen |

## <b>12.3 <font color='red'>|</font>     Set
---


Sets werden verwendet, um mehrere Elemente in einer einzigen Variablen zu speichern. Set ist einer von 4 integrierten Datentypen in Python, die zum Speichern von Datensammlungen verwendet werden, die anderen 3 sind List, Tupelund Dictionary, alle mit unterschiedlichen Qualit√§ten und Verwendungen.

Ein Set ist eine Sammlung, die sowohl **ungeordnet** als auch  **unver√§nderlich** ist. Ungeordnet bedeutet, dass die Artikel in einem Satz keine definierte Reihenfolge haben. Sets d√ºrfen nicht zwei Elemente mit demselben Wert haben. Sets sind unver√§nderlich, was bedeutet, dass die Elemente nicht ge√§ndert werden k√∂nnen, nachdem der Set erstellt wurde, aber man kann Elemente entfernen und neue Elemente hinzuf√ºgen.

| Funktion/Methode       | F/M | Beschreibung                                                      |
|------------------------|-----|-------------------------------------------------------------------|
| `set()`                | F   | Erzeugt ein neues Set aus einer iterierbaren Datenquelle.         |
| `len(s)`               | F   | Gibt die Anzahl der Elemente im Set zur√ºck.                        |
| `s.add(elem)`          | M   | F√ºgt ein Element zum Set hinzu.                                    |
| `s.remove(elem)`       | M   | Entfernt ein Element aus dem Set. Wirft einen Fehler, wenn das Element nicht vorhanden ist.  |
| `s.discard(elem)`      | M   | Entfernt ein Element, wenn es im Set ist. Ohne Fehlermeldung, falls nicht vorhanden.         |
| `s.pop()`              | M   | Entfernt und gibt ein zuf√§lliges Element aus dem Set zur√ºck.       |
| `s.clear()`            | M   | Entfernt alle Elemente aus dem Set.                                |
| `s.union(other_set)`   | M   | Gibt ein neues Set zur√ºck, das die Vereinigung zweier Sets ist.    |
| `s.intersection(other_set)` | M | Gibt ein neues Set zur√ºck, das die Schnittmenge zweier Sets ist.  |
| `s.difference(other_set)`   | M | Gibt ein neues Set zur√ºck, das die Differenz zweier Sets ist.     |
| `s.symmetric_difference(other_set)` | M | Gibt die symmetrische Differenz von zwei Sets zur√ºck.           |
| `s.update(other_set)`  | M   | Aktualisiert das Set mit der Vereinigung eines anderen Sets.       |
| `elem in s`            | F   | √úberpr√ºft, ob ein Element im Set vorhanden ist.                    |
| `for elem in s`        | F   | Iteriert √ºber jedes Element im Set.                                |





Sets werden mit geschweiften Klammern geschrieben.

**Besonderheit**: <br>
Auf Sets k√∂nnen die Operatoren der Mengenlehre angewandt werden (Vereinigungsmenge, Schnittmenge, Differenzmenge, symetrische Differenzmenge).

`````python
# Beispiel: Drucken Sie jede Frucht in einer Fruchtliste aus
A = {"Apfel", "Banane", "Kirsche", "Birne"}
B = {"Feige", "Aprikose", "Kirsche", "Birne", "Kiwi", "Orange"}

print(type(A), A)
print(type(B), B)
print('-' * 20)

print("Vereinigungsmenge: ", A | B)
print("Schnittmenge: ", A & B)
print("Differenzmenge: ", A - B)
print("symetrische Differenzmenge: ", A ^ B)
print('-' * 20)

# - anf√ºgen, l√∂schen, alle Elemente l√∂schen
A.add('Orange')       # anf√ºgen
A.remove('Birne')     # l√∂schen
B.clear()             # alle Elemete l√∂schen


print(type(A), A)
print(type(B), B)
`````

Variablen, die au√üerhalb einer Funktion erstellt werden (wie in den Beispielen unter), werden als globale Variablen bezeichnet. Globale Variablen k√∂nnen im Code verwendet werden, sowohl innerhalb als auch au√üerhalb von Funktionen. Wird eine Variable mit demselben Namen innerhalb einer Funktion erstellen, ist diese Variable lokal und kann nur innerhalb der Funktion verwendet werden. Die globale Variable mit dem gleichen Namen bleibt unver√§ndert, global und mit dem urspr√ºnglichen Wert.

Um eine globale Variable innerhalb einer Funktion zu erstellen, kann man das Schl√ºsselwort "global" verwenden.

`````python
# Beispiel : globale Variable  1
x = "globale Variable"

def myfunc():
  print("Dies ist eine " + x)

myfunc()
`````

`````python
# Beispiel : globale Variable 2
x = "globale Variable"

def myfunc():
    x = "lokale Variable"
    print("Dies ist eine " + x)

myfunc()
print("Dies ist eine " + x)
`````

`````python
# Beispiel : globale Variable 3
x = "globale Variable"

def myfunc():
    global x
    x = "globale Variable in einer Funktion"
    print("Dies ist eine " + x)

myfunc()
print("Dies ist eine " + x)
`````

## <b>15.2 <font color='red'>|</font>     Modul als Programm
---


Eine Code-Block grunds√§tzlich auf 2 Arten verwendet werden kann.

Er kann:

+ als eigenst√§ndiges Programm ablaufen oder
+ als Funktionen (gilt auch f√ºr Klassen oder Konstanten) in ein anderes Programm importieren werden.

Bei Aufruf des Code-Blocks vergibt Python einen internen Namen. Wird der Code-Block als eigenst√§ndiges Programm aufgerufen, ist dieser Name ``__main__``. Wird der Code-Block als Funktion aufgerufen, dann ist der Name ungleich ``__main__``. Der nach der Anweisung ``if __name__ == __main__:`` folgende Code wird nicht ausgef√ºhrt.

Durch die Verwendung von ``if __name_ == _main__``muss der Code-Block nicht in 2 Dateien gedoppelt werden (eigenst√§ndiges Programm plus Funktion). Beide Verwendungen k√∂nnen mit Hilfe von ``if __name__ == ‚Äù__main__`` durch eine einzige Datei bedient werden.

`````python
def fibonacci(n):
    x = 0
    y = 1
    for i in range(n):
        x, y = y, x + y
    return x
`````

`````python
# Beispiel: Aufruf als Funktion mit direkter Parameter√ºbergbae
#
fibonacci(3)
`````

`````python
# Beispiel: Aufruf als eigenst√§ndiges Programm mit Abfrage des Parameters aus dem Hauptprogramm  __main__
if __name__ == "__main__":
    a = input("Zahl: ")
    n = int(a)
    print(fibonacci(n))
`````

`````python
# Variante 2: Je Fehlerfall werden Anweisungen definiert
#
try:
    x = int(input("Bitte erste Zahlen eingeben: "))
    y = int(input("Bitte zweite Zahlen eingeben: "))
    z = x/y
    print(z)
    print("Programm ohne Fehler beendet")
except ZeroDivisionError:
    print("Fehler bei Division durch Null, Programm wird beendet")
except ValueError:
    print("Es sollte eine Zahl eingegeben werden")
except:
    print("Es gab einen anderen Fehler")
`````

`````python
# Variante 3: F√ºr mehrere Fehlerf√§lle wird ein Code-Block verwendet
#
def programm():
    x = int(input("Bitte erste Zahlen eingeben: "))
    y = int(input("Bitte zweite Zahlen eingeben: "))
    z = x/y
    print(z)
    print("Programm ohne Fehler beendet")

try:
    programm()
except (ZeroDivisionError, ValueError) as err:
    print("Es gab den Fehler", err)
except:
    print("Es gab einen weiteren Fehler")
`````

`````python
# Variante 4: Fehlerf√§lle werden nicht unterschieden (one-fits-all)
#
def programm():
    x = int(input("Bitte erste Zahlen eingeben: "))
    y = int(input("Bitte zweite Zahlen eingeben: "))
    z = x/y
    print(z)
    print("Programm ohne Fehler beendet")

try:
    programm()
except Exception as e:
    print("Mit Fehler: ", e, "beendet")
`````
